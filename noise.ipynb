```julia
fileName = "noise.log"
name = "noise"
data = readdlm(fileName,' ',skipstart=1)

workTimeInd = 1:size(data)[1];
workChanInd = [1;5:6;9:10]
```

```julia
time1 = data[:,workChanInd[1]];
spec = data[:,workChanInd[2:end]];

# plot(time,spec[:,1].-mean(spec[:,1]))
# plot!(time,spec[:,1]-spec[:,2])
```

## Темновые шумы в каналах 1 и 2
Как можно видеть из рисунка ниже наблюдается хорошая корреляция между  темновыми шумами в каналах 1 и 2.

```julia
plot(time1[1:100],spec[1:100,1],xlabel = "t, sec",ylabel = "U, ADC code")
plot!(time1[1:100],spec[1:100,2])
```

Коэффециент корреляции для данных каналов составляет: 

```julia
cor(spec[:,1], spec[:,2])
```

В отсутсвии корреляции даннй коэффециент должен сравняться нулю, а максимальное занчение составляет 1. 

## Шумы в каналах 1 и 2 в присутсвии света
Как можно видеть из следующего рисунка хорошая корреляция наблюдается также и  между шумами световых сигналов в каналах 1 и 2.

```julia
plot(time1[1:100],spec[1:100,3].-mean(spec[1:100,3]),xlabel = "t, sec",ylabel = "\$ U - \\overline{U},  ADC code\$")
plot!(time1[1:100],spec[1:100,4].-mean(spec[1:100,4]))
```

Коэффециент корреляции для световых сигналов составляет:

```julia
cor(spec[:,3], spec[:,4])
```
